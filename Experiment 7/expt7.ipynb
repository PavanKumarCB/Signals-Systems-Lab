## Reading file from Google drive

```python
import librosa #librosa is a Python package for music and audio processing
import librosa.display
import IPython.display as ipd #to play the audio signal
import matplotlib.pyplot as plt #plot the audio signal in time domain
ipd.Audio('/content/drive/MyDrive/Colab Notebooks/440Hz_44100Hz_16bit_05sec.wav')
#load a local WAV file
audio_path=('/content/drive/MyDrive/Colab Notebooks/440Hz_44100Hz_16bit_05sec.wav')
x , sr = librosa.load(audio_path)
```

## Getting the duration of audio

```python
ipd.Audio('/content/drive/MyDrive/Colab Notebooks/440Hz_44100Hz_16bit_05sec.wav')
#Duration of audio file
y=len(x) #number of samples x print(y)
#print(len(x))
y1=sr #sampling frequency sr
print(y1)
duration_of_sound=y/y1 #number of samples/sampling frequency
print(duration_of_sound,"second")
```

## Plotting the signal

```python
plt.figure(figsize=(14, 5))
#variable to change the x and y axix range
plt.grid(True )
librosa.display.waveshow(x)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

## Spectogram representation

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(20, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

## Mel Frequency Cepstral Coefficients

```python
librosa.load('/content/drive/MyDrive/Colab Notebooks/440Hz_44100Hz_16bit_05sec.wav')
mfccs = librosa.feature.mfcc(y=x, sr=sr, n_mfcc=100)
plt.figure(figsize=(10,4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title("'MFCC")
plt.tight_layout()
plt.show()
```

## Chroma

```python
x, sr = librosa.load('/content/drive/MyDrive/Colab Notebooks/440Hz_44100Hz_16bit_05sec.wav')
hop_length = 512
S = abs(librosa.stft(x))
chroma = librosa.feature.chroma_stft(S=S, sr=sr)
librosa.display.specshow(chroma, x_axis='time', y_axis='chroma', hop_length=hop_length, cmap='coolwarm')
```

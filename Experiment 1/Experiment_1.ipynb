```python
from numpy import *
import matplotlib.pyplot as plt
```

## 1. Unit step function

```python
t1 = linspace(-5,5,1000)
N = len(t1)
y1 = zeros(N)
for i in range(0,N):
  if(t1[i]) >= 0:
    y1[i] = 1

plt.plot(t1, y1)
plt.grid()
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Unit step function")
```

## 2. Unit Impulse function

```python
t2 = arange(-4,4)
y2 = arange(-4,4)

for i in list(y2):
  y2[i] = 0

for j in list(t2):
  if(t2[j] == 0):
    y2[j] = 1

plt.stem(t2,y2)
plt.grid()
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Unit impulse function")
```

## 3. Ramp signal

```python
t3 = arange(0,10)
y3 = [k for k in list(t3)]
plt.stem(t3,y3)
plt.grid()
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Ramp signal")
```

## 4. sinc signal

```python
t4 = arange(-2*pi, 2*pi, 0.01)
y4 = sinc(t4)
plt.plot(t4, y4)
plt.grid()
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("sinc signal")
```

## 5. Gaussian pulse

```python
t5 = arange(-0.5,0.5,0.01)
y5 = 1/(sqrt(2*pi)) * (exp(-t5**2/(2*0.01)))
plt.plot(t5, y5)
plt.grid()
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Gaussian pulse")
```
